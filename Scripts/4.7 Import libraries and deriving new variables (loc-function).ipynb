## Q1 - Creating the "price_label" and "busiest_day" columns

```python
# define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

## If the value within the “prices” column within the given row is less than or equal to 5, then return the string “Low-range product.”

## Or else, if the value within the “prices” column within the given row is greater than 5 and less than or equal to 15, then return the string “Mid-range product.”

## Or else, if the value of the “prices” column within the given row is greater than 15, then return the string “High-range product.”

## Or else, return the string “Not enough data.”

```python
# apply the function using the loc() function and create the price_label column
df_orders_products_large.loc[df_orders_products_large['prices'] > 15, 'price_range_loc'] = 'High range product'
```

## If the values of the prices column of the df_orders_products_large dataframe are greater than 15, then return the string "High range product"

```python
df_orders_products_large.loc[(df_orders_products_large['prices'] <= 15) & (df_orders_products_large['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

## If the values of the prices column of the df_orders_products_large_dataframe are less than or equal to 15 and greated than 5, then return the string "Mid range product"

```python
df_orders_products_large.loc[df_orders_products_large['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

## 0 = Saturday (Busiest)
## 1 = Sunday (2nd busiest)
## 2 = Monday
## 3 = Tuesday (2nd slowest)
## 4 = Wednesday (Slowest)
## 5 = Thursday
## 6 = Friday

```python
# create busiest days column using for-loops function
result = []

for value in df_orders_products_large["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest days")
  elif value == 1:
    result.append("Busiest days")
  elif value == 3:
    result.append("Slowest days")
  elif value == 4:
    result.append("Slowest days")
  else:
    result.append("Regular days")
```
