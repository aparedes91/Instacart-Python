## Q4

## Flag criteria:
## If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
## If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
## If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

## Three-step process:
## 1. Split the data into groups based on the “user_id” column.
## 2. Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
## 3. Create a new column, “max_order,” into which you’ll place the results of your aggregation.

```python
# create loyalty flag using transform() function
orders_products_merged_columns['max_order'] = orders_products_merged_columns.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
orders_products_merged_columns.head(30)
```

Note: A new column “max_order” has been created. We can now create another new column that assigns a loyalty flag to each customer according to the criteria via the loc() function. 

```python
# create loyalty flag using loc() function
orders_products_merged_columns.loc[orders_products_merged_columns['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
orders_products_merged_columns.loc[(orders_products_merged_columns['max_order'] <= 40) & (orders_products_merged_columns['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
orders_products_merged_columns.loc[orders_products_merged_columns['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# find the frequency of "loyalty_flag" column
orders_products_merged_columns['loyalty_flag'].value_counts(dropna = False)
```

## Q6

## Flag criteria:
## If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
## If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

## Three-step process:
## 1. Split the data into groups based on the “user_id” column.
## 2. Apply the transform() function on the “prices” column to generate the average prices for each user.
## 3. Create a new column, “max_spender,” into which you’ll place the results of your aggregation.

```python
# create a new column containing the mean frequency of the "prices" column
orders_products_merged_columns['max_spender'] = orders_products_merged_columns.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
orders_products_merged_columns.head(20)
```

Note: A new column “max_spender” has been created. We can now create another new column that assigns a spending flag to each user according to the criteria via the loc() function.

```python
# create a flag that assigns a "spending" label to a user ID based on its corresponding max_spender value
orders_products_merged_columns.loc[orders_products_merged_columns['max_spender'] < 10, 'spending_flag'] = 'Low spender'
```

```python
orders_products_merged_columns.loc[orders_products_merged_columns['max_spender'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# print the frequency of your new “spending_flag” column
orders_products_merged_columns['spending_flag'].value_counts(dropna = False)
```

Note: There are more low spenders than high spenders which indicates customers are more likely to spend less than more on products via Instacart.

```python
orders_products_merged_columns[['user_id', 'spending_flag', 'prices']].head(100)
```

## Q7

## Flag criteria
## If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
## If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
## If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

## Three-step process:
## 1. Split the data into groups based on the “user_id” column.
## 2. Apply the transform() function on the “days_since_prior_order” column to generate the median values for each user.
## 3. Create a new column, “ordering_behavior,” into which you’ll place the results of your aggregation.

```python
# create a new column containing the median frequency of the "days_since_prior_order" column
orders_products_merged_columns['ordering_behavior'] = orders_products_merged_columns.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
orders_products_merged_columns.head(20)
```

Note: A new column “ordering_behavior” has been created. We can now create another new column that assigns an order frequency flag to each user according to the criteria via the loc() function.

```python
# create a flag that assigns a "spending" label to a user ID based on its corresponding max_spender value
orders_products_merged_columns.loc[orders_products_merged_columns['ordering_behavior'] > 20, 'order_frequency_flag'] = 'Non frequent customer'
```

```python
orders_products_merged_columns.loc[(orders_products_merged_columns['ordering_behavior'] <= 20) & (orders_products_merged_columns['ordering_behavior'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
orders_products_merged_columns.loc[orders_products_merged_columns['ordering_behavior'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# print the frequency of your new “order_frequency_flag” column
orders_products_merged_columns['order_frequency_flag'].value_counts(dropna = False)
```

Note: It is recommended to send relevant notifications to frequent customers in case they are interested in buying the same item again.
