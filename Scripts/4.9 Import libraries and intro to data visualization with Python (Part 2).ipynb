## Sampling data

```python
# create seed and list holding True or False values to test the np.random.rant() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(orders_products_customers_merged)) <= 0.7
```

```python
dev
```

```python
# Store 70% of the sample in the dataframe big
big = orders_products_customers_merged[dev]
```

```python
# Store 30% of the sample in the dataframe small
small = orders_products_customers_merged[~dev]
```

```python
# check new sample subsets
len(orders_products_customers_merged)
```

```python
len(big) + len(small)
```

```python
# create a subset for line chart
df_small = small[['orders_day_of_week', 'prices']]
```

```python
# create line chart for the 'orders_day_of_week' and 'prices' column
line_chart_odow_prices = sns.lineplot(data = df_small, x = 'orders_day_of_week',y = 'prices')
```
